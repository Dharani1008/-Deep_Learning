```python
from PIL import Image
import glob
```

```python
image_list=[]
resized_images=[]
```

```python
for filename in glob.glob('/home/brillersys/Documents/flower/*.jpg'):
    print(filename)
    img = Image.open(filename)
    image_list.append(img)
```

```python
for image in image_list:
    image = image.resize((150,150))
 
    resized_images.append(image)
```

```python
for (i,new) in enumerate(resized_images):
    new.save('{}{}{}'.format('/home/brillersys/Documents/img150/img',i+1,'.jpg'))
```
